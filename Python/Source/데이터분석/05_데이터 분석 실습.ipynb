# 정규표현식을 활용한 데이터 분석
    + 정규표현식이란?
        - 특정한 규칙을 가진 문자열을 표현하기 위해 사용하는 형식
        - 주로 문자열의 검색 및 치환에 활용
        - re 모듈을 총해서 구현
        - complie() 함수를 사용하여 정규표현식 내용을 적용

    다양한 규칙이 존재함

````python
"""
    - 정규표현식의 예시 
             . . : 1개 문자 와 일치
             . [ ] : [,] 사이의 문자중 하나 선택
             > [abc]d : ab,bd,cd 포함
             > [a-zA-Z] : 알파벳 모두 포함
             > [0-9] : 숫자 를 모두 포함
             . * : 0 개 이상의 문자 포함
             > a*b : b,aab,ab,aaaaab 등
             . {m,n} : m회 이상 n 회 이하
             > a{1,3}b -> ab,aab,aaab 포함 ( a 의 횟수 를 1 ~ 3개 를 표현 가능함)
             . 그 외 
             > ?, +, (), ^ 등의 다양한 규칙이 존재함
"""
````

````python
# 정규식 사용을 위한 모듈 import
import re

# 정규식 패턴 정의 
pat = 'apple' # 특정 문자열이 나타내는 패턴

# 패턴을 기준으로 문자열에서 찾아내기
text = 'I have an apple and banana'

result = re.findall(pat, text) # 샘플에서 특정한 패턴을 찾아네는 기능

print(result)
````

* 콤마를 제거한 테이터만 추출하기

````python
import re
a = "1,2,3,4,5"

pattern = re.compile('[^0-9]') # 숫자가 아닌것들을 제거
result = re.sub(pattern, '', a) # 원본 a 문자열에서 pattern과 일피하는 것을 ''과 바꿔라

print(result)
````

* 이메일 주소 찾기 (특정 데이터에서 메일주소 부분만 발췌)

* 웹페이지에서 데이터를 가지고 와서 모든 HTML 태그를 제거하고 표현하기

* 사이트 주소만 발췌

````python
c = '''<div class="select_div">
    <span data-i18n="footer.relationDites">연관사이트</span>
    <ul>

    </ul>
</div>
<ul class="list_sns">
    <li><a href="http://www.facebook.com/SmartEducation/" target="_blank"><img src="/res/ims/common/btn footer sns1.png" alt="페이스북"</a></li>
    <li><a href="http://blog.naver.com/Samart-edu/" target="_blank"><img src="/res/ims/common/btn footer sns2.png" alt="네이버 블로그"</a></li>
    <li><a href="http://www.facebook.com/SmartEducation/" target="_blank"><img src="/res/ims/common/btn footer sns3.png" alt="유투브"</a></li>
</ul>'''
````

````python
import re

pattern = re.compile('http://[^\s"\'<>]+')
# [^\s"\'<>]+ : ^ 는 부정
#              \s 는 공백 문자
#              "'<> 는 쌍따옴표, 홑따옴표, <, > 문자를 포함
#              + 는 패턴이 하나 이상 반복되는 것일 것

result = pattern.findall(c)
for url in result :
    print(url)
````

````python
# 2. 새로운 리스트에 각 영화의 장르를 분리
import csv
f = open('movies.csv', 'r' , encoding = 'utf-8')
data = csv.reader(f)
next(data)
result = [] #모든 장르를 담을 영화 장르 리스트
for i in data :
    result.extend(i[2].split('|'))
print(len(result))
print(result)
````

````python
# 3. 중복된 장르명칭을 제외하고 조회하기 (장르 종류 확인)
genre = set(result)
print(len(genre))
print(genre)
````

````python
# 4. 장르별 영화의 개수 확인
# > 사전을 이용한 패턴, 해당 장르가 사전에 키로 존재할 경우 +1, 존재하지 않을 경우 1

import csv
f = open('movies.csv', 'r' , encoding = 'utf-8')
data = csv.reader(f)
next(data)
result = [] #모든 장르를 담을 영화 장르 리스트
for i in data :
    result.extend(i[2].split('|'))

# 생성된 리스트를 반복하여 추출 후 비교하여 사전에 등록
count = {} # 장르를 키로 가지고 빈도수를 값으로 가지는 사전
for i in result :
    if count.get(i) :  # 장르를 키로 가지고 있는 사전데이터가 있다면 
        count[i] +=1 # 해당 장르 +1
    else : 
        count[i] = 1
print(count)
````

````python
# 1. 위의 예제에서 장르 가 가장 다양한 영화를 찾아 보세요 (최대 장르 개수 가진 영화 모두 구하기)
import csv
f = open('movies.csv', 'r' , encoding = 'utf-8')
data = csv.reader(f)
next(data)
title = {}
for i in data :
    if title.get(i[1]) :
        count = i[2].split('|')
        title[i[1]] += len(count) 
    else :
        count = i[2].split('|')
        title[i[1]] = len(count) 

mostG = max(title.values())
print(mostG)
for i in title :
    if title[i] == mostG:
        print(i)
````

````python
# 교수님 답안
import csv
f = open('movies.csv', 'r' , encoding = 'utf-8')
data = csv.reader(f)
next(data)

moviePergenre = {}
for i in data :
    moviePergenre[i[1].split('(')[0]] = len(i[2].split('|'))


mmaxGenreCount = max(moviePergenre.values())

resultList = [key for key.value in moviePergenre.items() if value == mmaxGenreCount ]

for title in resultList :
    print(title, moviePergenre[title])
````

````python
# 2. 연도별 가장 많이 개봉된 영화의 장르 를 찾아 보세요
import csv
f = open('movies.csv', 'r' , encoding = 'utf-8')
data = csv.reader(f)
next(data)

yeargenre = {}
genre = {}
for i in data :
    year =  i[1][-5:-1]
    genre = i[2].split('|')
    if year not in yeargenre :
        yeargenre[year] = {}
    for j in genre :
        if yeargenre[year].get(j) :
            yeargenre[year][j] += 1
        else :
            yeargenre[year][j] = 1
print(yeargenre)
````

````python
# 교수님 답안
import csv
f = open('movies.csv', 'r' , encoding = 'utf-8')
data = csv.reader(f)
next(data)

mpg = {}
for i in data :
    # 연도
    year =  i[1][-5:-1]
    genre = i[2].split('|')

    if year not in mpg :
        mpg[year] = {}
    for g in genre :
        if g not in mpg[year] :
            mpg[year][g] = 1
        else :
            mpg[year][g] += 1
    


for year,genre in mpg.items() :
    maxg = max(genre, key = genre.get)
    
    print(f"{year}년도에는 {maxg}장르가 {genre[maxg]}회로 가장 많이 상영되었습니다." )
````

# 부산광역시의 해당 기간 동안 월별 행정구역별 주민등록 인구 및 세대 현황 구하기

* 데이터 파악
    + 총 인구수는 두번째 열부터 시작하여 다음월의 인구수는 6칸 뒷열에 반복적으로 나타난다.
    + 행정구역별 인구 수는 세번째 행 부터 확인할 수 있다.

```python
import pandas as pd
result = pd.read_csv('202109_202309_주민등록인구및세대현황_월간.csv', encoding='cp949')
print(result)
```

```python
# 행정 구역의 명칭
print(result.iloc[1,0]) # 1 행의 0 번 컬럼의 데이터
```

```python
# 첫번째 행정구역의 첫월 총 인구수 
print(result.iloc[1,1])
```

```python
# 1행의 모든 데이터 출력
print(result.iloc[1])
```

```python
# 1행 1열 (첫 행정구역의 첫월 총인구수) ~ 마지막 데이터 까지 6칸씩 열을 띄워서 표현
print(result.iloc[1, range(1, len(result.columns), 6)])
```

```python
# 1행 0열의 데이터 (행정구역 이름)
print(result.iloc[1,0])
# 1행 1열 (첫 행정구역의 첫월 총인구수) ~ 마지막 데이터 까지 6칸씩 열을 띄워서 표현
print(result.iloc[1, range(1, len(result.columns), 6)])
```

```python
# 반복문을 통하여 모든 행정구역의 월별 인구 추이 구하기
import pandas as pd
result = pd.read_csv('202109_202309_주민등록인구및세대현황_월간.csv', encoding='cp949')

# 행정 구역 별 인구의 데이터가 들어갈 사전
data = {}
for i in range(1, len(result)) : # 1행 ~ 총 갯수 - 1행
    # 사전에 지역구를 key로 담기
    data[result.iloc[i,0]]= result.iloc[1,range(1,len(result.columns), 6)]
print(data)
```
